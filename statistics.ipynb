# variance

```python
import numpy as np

data =[2,4,6,7,5,6,5,9,4]

sample_variance = np.var(data, ddof = 1) # 1 hai ya ddof(dalta degree of freedom) agr iski value 1 hain to sample variace or agr value 0 hai to population variannce

population_variance = np.var(data,ddof = 0)

print("sampale variance is : ",sample_variance)
print("population variance is : ",population_variance)
```

# Data Representations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Create a figure and axes
fig, axs = plt.subplots(3, 2, figsize=(14,12)) 
fig.suptitle('Histogram with KDE of various Distribytaion', fontsize=16)

# Normal Distribution (Gaussion Distributaion)
data_normal = np.random.normal(loc=0, scale=1, size=1000) # mean=0 std dev=1
sns.histplot(data_normal, kde=True, ax=axs[0,0], color='blue', stat='density',bins=20)
axs[0,0].set_title('Noramal Distribution')
axs[0,0].grid(True)

#Skewed Distribution
data_right_skewed = np.random.exponential(scale=1, size=1000) # right skewed
data_left_skewed = -np.random.exponential(scale=1, size=1000) # left skewed

#right skewed histogram
sns.histplot(data_right_skewed, kde=True, ax=axs[0,1], color='red', stat='density', bins=30, label='Right skewed')
#left skewed istogram
sns.histplot(data_left_skewed, kde=True, ax=axs[0,1], color='green', stat='density', bins=30, label='left skewed')
axs[0,1].set_title('Skewed Distribution')
axs[0,1].legend()
axs[0,1].grid(True)

#uniform Distribution 
data_uniform = np.random.uniform(low=0, high=1, size=1000) # uniform distribution between 0  and  1
sns.histplot(data_uniform, kde=True, ax=axs[1,0], color='purple', stat='density', bins=30)
axs[0,1].set_title('Uniform Distribution')
axs[0,1].grid(True)

# Bimodel Distribution
data_bimodal = np.concatenate([np.random.normal(loc=-2, scale=1, size=500),np.random.normal(loc=2, scale=1, size=500)])
sns.histplot(data_bimodal, kde=True, ax=axs[1,1], color='Orange', stat='density', bins=30)
axs[1,1].set_title('Bimodel Distribution')
axs[1,1].grid(True)

#multimodel Distribution

data_multimodel = np.concatenate([
    np.random.normal(loc=-6, scale=1.5, size=500),
    np.random.normal(loc=0, scale=1, size=500),
    np.random.normal(loc=6, scale=1.5, size=500)
])

sns.histplot(data_multimodel, kde=True, ax=axs[2,0], color='brown', stat='density', bins=30)
axs[2,0].set_title('Multimodel Distribution')
axs[2,0].grid(True)

#hide empty subplot
axs[2,1].axis('off')

#adjust layout
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

# T Test & P Test

```python
import numpy as np
from scipy import stats

#step 1 : create tow random sample datasets
#suppose  these are test scores from two group(e.g, group A and B)
group_A = np.random.normal(75,10,30) #group a with  mean 75, std 10, 30 samples
group_B = np.random.normal(80,12,30) #group a with  mean 80, std 12, 30 samples

#step 2: Perform on independent t_test
t_statistics, p_value = stats.ttest_ind(group_A,group_B)

#step 3: Display the p-value and t-statistics
print(f"T-statistics : {t_statistics}")
print(f"P-value : {p_value} ")

#Step : Interpretion based on p-value
alpha = 0.05 # significance level
if p_value <= alpha:
    print("reject the null hypothesis (Significant diffrence between groups)")
else:
    print("Fail to reject the null hypothesis (No significant diffrence between group)")
```

# Z value Test

##### Imagine a company claims that the average weight of their product is 500 grams. We want to test if this claim is true by collecting a sample of 30 products. The sample mean is 505 grams, and the population standard deviation is 10 grams. We will conduct a Z-test to see if the difference in the mean is significant or not.

```python
import numpy as np
from scipy import stats

#step 1 : Define the know parameters
population_mean = 500 #claimed the avrage (mu)
sample_mean = 505 # sample mean 
population_std = 10 # population standard deviation (sigma)
sample_size = 30 # sample size (n)

#step 2 : calculate the z-value 
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
print(f"Z-value : {z_value}")

#step 3 : calculate the  p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_value)))
print(f"p-value : {p_value}")

#step 4 : Set significanc level
alpha = 0.05

#step 5 : compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis (significant difference)")
else:
    print("Fail to reject the null hypothesis (no significant diffrence)")
```
